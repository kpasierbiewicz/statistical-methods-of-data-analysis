```python
import pandas as pd
```

```python
df = pd.read_csv('yeast.data', delimiter = "\t", header=None, names=["mcg", "gvh", "alm", "mit", "erl", "pox", "vac", "nuc", 'class'])
df
```

```python
df['class'] = pd.factorize(df['class'])[0] + 1
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn import datasets, linear_model
from sklearn.neighbors import KNeighborsClassifier
```

```python
y = df['class']
X_train, X_test, y_train, y_test = train_test_split(df, y)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
for i in range(3, 7):
    kf =KFold(n_splits=i, random_state = None)
    classifier = KNeighborsClassifier(n_neighbors = 5)
    score = cross_val_score(classifier, df, y, cv = kf)
    print(score)
    print(score.mean())
```

```python
from sklearn import tree
X = df
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
tree.plot_tree(clf)
```
