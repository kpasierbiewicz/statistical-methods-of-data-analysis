###  II. Proszę uzupełnić brakujące elementy korzystając ze strategii najbliższego sąsiada

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
df = data.copy()
df.drop('date', axis = 1, inplace = True)
imputed = imputer.fit_transform(df)
df_imputed = pd.DataFrame(imputed, columns=df.columns)
df_imputed
```

### III.	Korzystając z jednowymiarowej metody wykrywania elementów odosobnionych proszę wyznaczyć odstające przypadki – dla każdej z cech z osobna.
- dla przypomnienia w tej metodzie elementami odstającymi są przypadki znajdujące się poza przedziałem średnia +/- 3 odchylenia standardowe.
W przypadku których elementów uzyskujemy interesujące informacje nt. danych a kiedy mamy do czynienia z „oczywistą oczywistością”?


```python
sigma = df_imputed.std()
srednia = df_imputed.mean()
```

```python
sigma
```

```python
srednia
```

```python
low_boundary = (srednia - 3 * sigma)
upp_boundary = (srednia + 3 * sigma)
num_of_outliers_L = (df_imputed < low_boundary).sum()
num_of_outliers_U = (df_imputed > upp_boundary).sum()
outliers_3sigma = pd.DataFrame({'lower_boundary':low_boundary, 'upper_boundary':upp_boundary,'num_of_outliers_L':num_of_outliers_L, 'num_of_outliers_U':num_of_outliers_U})
outliers_3sigma
```

###  IV. Proszę zrealizować wielowymiarowe wykrywanie elementów odstających z użyciem metody Local Outlier Factor

dla przypomnienia w tej metodzie elementami odstającymi są przypadki znajdujące się poza przedziałem średnia +/- 3 odchylenia standardowe. W przypadku których elementów uzyskujemy interesujące informacje nt. danych a kiedy mamy do czynienia z „oczywistą oczywistością”? IV. Proszę zrealizować wielowymiarowe wykrywanie elementów odstających z użyciem metody Local Outlier Factor (Python: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.LocalOutlierFactor.html ; MATLAB: https://www.mathworks.com/matlabcentral/fileexchange/72244-density-based-outlier-detection-algorithms; R: https://github.com/BlueBirdHouse/DDoutlier) 

```python
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors=2)
clf.fit_predict(df_imputed)
df_outlier = clf.negative_outlier_factor_
df_outlier
```
